```python
import re

def remove_numbers(text):
    text = re.sub(r'\d+', '', text)  # 숫자를 공백으로 대체
    return text.strip()  # 양쪽 공백 제거

# 'title' 열에서 숫자 제거
df['title'] = df['title'].apply(remove_numbers)
```

```python
from konlpy.tag import Okt

# Okt 객체 생성
okt = Okt()

# 형태소 분석 함수
def tokenize(text):
    tokens = okt.morphs(text)
    return tokens

# 'title' 열에 대해 형태소 분석 적용
df['title_tokens'] = df['title'].apply(tokenize)
```

```python
# 텍스트를 띄어쓰기로 토큰화하는 함수 정의
def tokenize_by_space(text):
    tokens = text.split()
    return tokens

# 'title' 열에 있는 텍스트를 띄어쓰기로 토큰화하여 새로운 열로 추가
df['title_tokens'] = df['title'].apply(tokenize_by_space)

```

```python
with open('stopword.txt', 'r', encoding='utf-8') as file:
    stopwords = file.read().splitlines()
```

```python
# 불용어 처리 함수
def remove_stopwords(tokens):
    filtered_tokens = [token for token in tokens if token not in stopwords]
    return filtered_tokens

# 'title_tokens' 열에 대해 불용어 처리 적용
df['processed_title_tokens'] = df['processed_title_tokens'].apply(remove_stopwords)
```

```python
def youngib(tokens):
    for i in range(len(tokens)-1):
        if (tokens[i] == '영') and (tokens[i+1].startswith('입')):
            tokens[i] = '영입'
            if len(tokens[i+1]) > 1:
                tokens[i+1] = tokens[i+1][1:]
            else:
                tokens.pop(i+1)
                
    return tokens
e = e.apply(youngib)
e
```

```python
from soynlp.tokenizer import MaxScoreTokenizer
import urllib.request
from soynlp import DoublespaceLineCorpus
from soynlp.word import WordExtractor
from soynlp.tokenizer import LTokenizer
```

```python
urllib.request.urlretrieve("https://raw.githubusercontent.com/lovit/soynlp/master/tutorials/2016-10-20.txt", filename="2016-10-20.txt")
```

```python
corpus = DoublespaceLineCorpus("2016-10-20.txt")
```

```python
word_extractor = WordExtractor(min_frequency=100,
    min_cohesion_forward=0.05, 
    min_right_branching_entropy=0.0
)
word_extractor.train(corpus)
word_score_table = word_extractor.extract()
```

```python
urllib.request.urlretrieve("https://raw.githubusercontent.com/lovit/soynlp/master/tutorials/doublespace_line_corpus_sample.txt", filename="doublespace_line_corpus_sample.txt")
```

```python
corpus = DoublespaceLineCorpus("2016-10-20.txt")
```

```python
word_extractor.train(corpus)
word_score_table = word_extractor.extract()
```

```python
scores = {word:score.cohesion_forward for word, score in word_score_table.items()}
```

```python
maxscore_tokenizer = MaxScoreTokenizer(scores=scores)
maxscore_tokenizer.tokenize('병역 기피 논란  드디어 입 열었다소속팀서 안 놔줘 정말 오해')
```

```python
l_tokenizer = LTokenizer(scores=scores)
l_tokenizer.tokenize("영입 제의를 받았다", flatten=False)
```
